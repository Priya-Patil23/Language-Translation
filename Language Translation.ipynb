```python
english_sentences = english["English words/sentences"].tolist()
french_sentences = french["French words/sentences"].tolist()
```

```python
english_sentences
```

```python
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input , Embedding , LSTM , Dense
from tensorflow.keras.models import Model
from tensorflow.keras.callbacks import EarlyStopping , ModelCheckpoint
from sklearn.model_selection import train_test_split
```

```python
english_tokenize = Tokenizer()
english_tokenize.fit_on_texts(english_sentences)
english_tokened_sent = english_tokenize.texts_to_sequences(english_sentences)

french_tokenize = Tokenizer()
french_tokenize.fit_on_texts(french_sentences)
french_tokened_sent = french_tokenize.texts_to_sequences(french_sentences)

eng_vocab = len(english_tokenize.word_index) + 1
french_vocab = len(french_tokenize.word_index) + 1

maxlength = max([len(i) for i in english_tokened_sent + french_tokened_sent])
eng_padded = pad_sequences(english_tokened_sent , maxlen=maxlength , padding = 'post')
french_padded = pad_sequences(french_tokened_sent , maxlen=maxlength , padding = 'post')
```

```python
# Encoder

encoder_input = Input(shape= (maxlength , ))
encoder_embedding = Embedding(eng_vocab , 128)(encoder_input)
encoder_lstm =LSTM(128 , return_state = True)
_ , state_h , state_c = encoder_lstm(encoder_embedding)

states = [state_h, state_c]

# decoder

decoder_input = Input(shape = (None , ))
decoder_embedding = Embedding(french_vocab , 128)(decoder_input)
decoder_lstm = LSTM(128 , return_state = True , return_sequences=True)
decoder_output , _ , _ = decoder_lstm(decoder_embedding , initial_state = states)
decoder_dense = Dense(french_vocab , activation="softmax")(decoder_output)

model = Model(inputs = [encoder_input , decoder_input] , outputs = decoder_dense)
```

```python
model.summary()
```

```python
model.compile(loss = 'sparse_categorical_crossentropy' , optimizer = "Adam" , metrics = ['accuracy'])
```

```python
xtrain , xtest , ytrain , ytest = train_test_split(eng_padded , french_padded , test_size=0.2 , random_state=42)
```

```python
xtest
```

```python
def early():
    return EarlyStopping(monitor = 'val_loss',patience = 5,restore_best_weights=True,min_delta = 0.001)
def point(path):
    return ModelCheckpoint(path,monitor = 'val_loss',save_best_only=True,save_weights_only=True,save_freq="epoch")
```

```python
history = model.fit([xtrain , xtrain] , ytrain , epochs = 10 , validation_data = ([xtest,xtest] , ytest),
          use_multiprocessing=True , callbacks = [early() , point("translate.ckpt")]  , batch_size = 64)
```

```python
model.load_weights("translate.ckpt")
model.save("translation_again.h5")
```

```python
def translate_sentence(sentence):
    seq = english_tokenize.texts_to_sequences([sentence])
    padded = pad_sequences(seq, maxlen=maxlength, padding='post')
    translated = np.argmax(model.predict([padded, padded]), axis=-1)
    
    translated_sentence = []
    for i in translated[0]:
        if i in french_tokenize.index_word:
            translated_sentence.append(french_tokenize.index_word[i])
        else:
            translated_sentence.append(' ')  
        
    return ' '.join(translated_sentence)

input_sentence = "hello I am Ajinkya"
translated_sentence = translate_sentence(input_sentence)
print(f"Input: {input_sentence}")
print(f"Translated: {translated_sentence}")
```
